# 05. AI Basic Math

### AI 학습을 위한 기초 수학
#### 함수
- 두 집합 사이의 `관계` 혹은 `규칙`을 나타냄
- `y = f(x)` : x는 입력 값, y는 출력 값
- 데이터 분석에서 함수는 x(feature)와 출력 y(label) 간의 관계를 설명하는데 사용
- 함수의 출력에 예측값을 나타낼 때는 `y^`와 같이 햇 모양(^)을 씌워 표현
- 1차 함수
	- `y = ax + b`
    - a는 기울기, 직선이 얼마나 가파른지 결정
    - b는 절편, 그래프가 y축과 만나는 지점
- 2차 함수
	- `y = a(x-p)^2 + q`
    - a : 그래프의 모양과 방향을 결정(양수면 아래로 볼록, 음수면 위로 볼록)
    - (p,q) : 꼭짓점의 좌표
- 미분
	- 함수의 `순간 변화율`을 구하는 것
    - 특정 점에서의 `기울기`를 계산
    - 두 점 사이 기울기 ➡️ **평균 변화율**
    - 미분(기울기) ➡️ **순간 변화율**
    - 꼭짓점에서 멀어질수록 기울기가 커진다
    - 머신러닝에서 **최솟값**을 찾는것이 목표 ➡️ 오차를 최소화
    - **데이터를 가장 잘 설명할 수 있는 가중치를 얻는 과정**
    - 최적화나 그래프의 증가/감소 경향을 분석할 수 있음
- 편미분
	- 변수가 여러 개인 다변수 함수에 대해 다른 변수들은 고정시키고, 특정 변수에 대해서만 변화율(기울기)을 구하는 것
    - 그라디언트 
     	- **기울기 벡터랑 동일**
     	- 최솟값을 찾는 길잡이
     	- 최솟값에서는 그라디언트가 0
     	- 다변수 함수의 모든 편미분을 모아놓은 벡터

# 04. Machine Learning Basic

### Machine Learning
#### 머신러닝의 과정
- 입출력 정의 → 모델 만들기 → 손실함수 정의 → Weight 최적화

#### 머신 러닝의 핵심요소
- 데이터
- 모델
- 학습 알고리즘
- 평가 및 예측

#### 데이터
- 머신러닝은 데이터를 기반으로 모델을 학습하고 예측
- 입력 데이터(features)와 출력 데이터(labels)로 구성
- 데이터 품질이 모델 성능에 큰 영향을 미치며, 전처리와 특징 추출이 필수적

#### 모델
- 데이터를 통해 패턴/관계를 학습하고 결과를 예측하는 수학적인 알고리즘의 집합 → 간단하게 `함수`를 의미

### 데이터 전처리 및 데이터 분할
#### 데이터 전처리
- 데이터 분석 및 모델링 이전에 데이터를 정리하고 품질을 개선하는 과정
- 데이터 사이언티스트들이 소모하는 전체 시간의 약 60%를 차지함
- 목적
	- 데이터 품질 향상
    - 분석/모델링 효율성 향상
    - 계산 성능 최적화

#### 데이터 분할
- 훈련 데이터, 검증 데이터, 테스트 데이터로 나눔
- 훈련 데이터로 학습하고, 검증 데이터로 성능을 평가, 테스트 데이터를 사용하여 최종 평가
- 독립성 유지 : 훈련, 검증, 데이터는 서로 겹치지 않아야함
- 목적
	- 일반화 성능 평가
    - 과적합(Overfitting) 검증
    - 하이퍼파라미터 튜닝
- 과적합 : 머신러닝 모델이 학습 데이터에 지나치게 맞춰져, 실제 데이터에서는 예측 성능이 급격히 떨어지는 현상을 의미
- 하이퍼파라미터 : `사용자가 직접 설정`하는 값, 모델의 구조나 학습 방식을 제어
- 파라미터 : `모델 내부에서 학습을 통해 결정`되는 값, 가중치 편향 계수등 데이터에 의해 산출


### 머신러닝의 학습 방법
- 지도 학습
- 비지도 학습
- 강화 학습

#### 지도 학습
- 입력 데이터와 이에 해당하는 정답(라벨)이 있는 데이터를 사용해 학습하는 방식
- 새로운 데이터가 주어졌을 때 적절한 결과를 예측
- 분석모형 : 회귀, 분류
- 활용 예시 : 이메일 스팸 분류, 질병 진단, 가격 예측

#### 비지도 학습
- 정답(라벨) 없이 입력 데이터만 가지고 학습하는 방식
- 데이터 안의 `패턴이나 구조를 발견`하거나, `비슷한 데이터끼리 그룹화`하는데 사용
- 분석모형 : 군집, 차원 축소
	- 군집화 : 비슷한 데이터끼리 그룹화 ex) 마케팅 고객 군집화
    - 차원 축소 : 여러 변수를 요약하여 적은 차원으로 변환 ex) 이미지 압축
- 활용 예시 : 고객 세분화, 데이터 시각화, 추천 시스템

#### 강화 학습
- `보상과 벌칙`을 기반으로 AI가 환경과 상호작용하며 최적의 전략을 학습
- `보상 중심`으로 순찾거 의사결정을 통해 탐험과 활용의 균형을 맞추며, 라벨 데이터 없이 환경과 상호작용하면서 학습

### 인공신경망
#### 인경신경 - 노드(Node)
- 인공 신경은 인공신경망을 이루는 작은단위
- 입력값을 받아서 간단한 계산을 하고 결과를 내보내는 역할을 함
- 바이어스(민감도)
	- 입력 변화에 따라 모델의 출력이 얼마나 빠르게 변하는지를 의미
    - 모델의 출력이 입력 변화에 얼마나 민감하게 반응하는지 결정
- 가중치
	- 입력값에 곱해져 출력에 영향을 주며 모델이 입력-출력 관계를 학습할 때 자동으로 조정
    - 입력과 출력을 연결하는 매개변수
- `y = wx + b` : w,b는 파라미터
- 머신러닝은 가중치와 민감도를 찾아내는 것 그리고 오차에 대한 민감도를 계산해서 업데이트하는 과정
- 두 요소는 모델의 성능과 해석에 직접적인 영향을 미치므로, 학습 과정에서 세밀하게 조정됨
- `Perceptron` : 단일 계산층
- `MLP(Multi Layer Perceptron)` : 다층 퍼셉트론, 인접한 두층의 뉴런 간에는 완전 연결
- **활성화 함수**
	- 노드는 활성화 함수를 가지고 있음
	- 노드에서 계산된 값을 기반으로 어떤 추력 값을 다음 층으로 보낼지 결정하는 함수
    - 종류 : Sigmoid, tanh, ReLU 등

# 07. Back Propagtion(역전파)

#### 역전파
- 경사하강법을 하면서 어떻게 가중치를 업데이트 하느냐?
- 오차를 계산, 신경망의 `가중치를 업데이트`하는 알고리즘
- 발생한 오차에 대해서 각각 w에 대한 기여도를 계산하여 w를 업데이트
- 체인룰
    - 계산의 핵심
    - 역전파는 `미분`을 이용함
    - 체인 룰을 활용하여 오차가 각 가중치에 미치는 영향을 계산
    - 입력층에 가까워지면서 계산이 깊어질수록 가중치는 0값에 가까워지므로 업데이트가 되지 않아 파악이 어렵다
- 주요 단계
    - 순전파 : 입력값 → 출력값 생성, 모델의 예측값과 실제값의 차이를 측정하여 오차를 계산, 값을 한번 통과 시켜줘야함
    - 역전파 : 오차를 `출력층에서 입력층 방향으로` 전파, 각 가중치가 오차에 얼마나 영향을 미쳤는지 계산, 기울기를 이용해 가중치를 업데이트

#### 분류
- 최적의 결정 경계선을 찾는 것
- 입력 데이터를 사전에 정의된 여러 클래스 중 하나로 구분하는 알고리즘
- 클래스
	- 이진 분류
    - 다항 분류
    - 다중 분류 : 샘플이 여러 클래스에 동시에 속할 수 있는 분류
- 선형 모델 : 로지스트 회귀

#### 로지스틱 회귀

- 로지스틱 회귀
    - 선형 회귀를 기반으로 한 분류 알고리즘, 입력 데이터를 분석하여 특정 클래스에 속할 확률을 예측
    - 계단 함수는 `미분이 불가능` ➡️ `경사하강법 적용할 수 ❌`
    - 시그모이드 함수로 계단 함수를 곡선 형태로 변형
    - 시그모이드 함수 : 입력값을 0과 1 사이의 값으로 변환하는 활성화 함수
    - 시그모이드 함수 단점 : 기울기 소실, 함수의 중심이 0이 아님
    - 최근 자주 사용하지 않음

- 로지스틱 회귀의 손실함수
    - Binary Cross-Entropy = 확률 기반 분류을 통해 도출된 손실 함수
    - 손실 미분 → 경사하강법 → 가중치 업데이트
    - 예측 확률이 실제값과 가까워질수록 손실 감소

#### 다항분류
- Softmax
    - 3개 이상의 Label을 갖는 데이터에 대한 분류 작업
    - One-Hot Encoding : 범주형 데이터를 숫자로 바꿀 때 사용하는 기법
    - Softmax 함수
        - 출력층 활성화 함수
        - `입력값을 확률로 변환`하는 함수
        - 출력 값들의 합이 항상 1이 되도록 정규화함
    - Softmax 함수 특징 : 확률 분포 생성, 해석 용이성(예측 확률 제공)

- 다항 분류 손실함수
    - 입력 데이터를 One-Hot encoding으로 변환
    - Softmax적용 ➡️ 확률로 변환
    - 손실 계산 (Cross-Entropy) : `log` 취해서 `-` 해줌 ➡️ 손실 함수
